```python
import pandas as pd
import numpy as np
```

```python
name_hla = 'HLA-A*01:01' ## here insert the HLA; for a multi-allelic model, use 'multi' ##
#name_hla = 'multi'

fn = 'covid_' + name_hla + '/covid_9mers_scored_YesReal_0_hu_5_l120.001_AL9.0_SA9_RWYesAlpha_TR1.txt'
scores_out = pd.read_csv(fn, sep = '\t')
fn2 = 'covid_9mers_gene.txt'
scores_out2 = pd.read_csv(fn2, sep = '\t')
```

```python
## Write ranked 9-mers for HLA*A01:01, for all proteins ## 

llt = scores_out['RBM Log-Lik'].values
seqs = list(scores_out['Sequences'].values)
prot = scores_out2['Gene'].values

Scores_list = zip(seqs, llt, prot)
Srank = sorted(Scores_list, key = lambda x : x[1], reverse = True)
data_ranked = {'Peptide':list(np.array(Srank)[:,0]),
        'RBM-MHC score': list(np.array(Srank)[:,1]),
        'Protein': list(np.array(Srank)[:,2]),
        }
pd.DataFrame (data_ranked, columns = ['Peptide','RBM-MHC score','Protein'])
```

```python
## Restrict the ranking to a chosen protein, here N ## 
## The symbols for proteins are found by : print(list(np.unique(prot)))

PROT = 'N'

llt = scores_out[scores_out2['Gene']==PROT]['RBM Log-Lik'].values
seqs = list(scores_out[scores_out2['Gene']==PROT]['Sequences'].values)
prot = list(scores_out2[scores_out2['Gene']==PROT]['Gene'].values)

Scores_list = zip(seqs, llt, prot)
Srank = sorted(Scores_list, key = lambda x : x[1], reverse = True)
data_ranked = {'Peptide':list(np.array(Srank)[:,0]),
        'RBM-MHC score': list(np.array(Srank)[:,1]),
        'Protein': list(np.array(Srank)[:,2]),
        }
pd.DataFrame(data_ranked, columns = ['Peptide','RBM-MHC score','Protein'])
```
